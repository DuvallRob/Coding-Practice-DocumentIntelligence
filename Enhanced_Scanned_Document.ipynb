```python
from google.colab.patches import cv2_imshow
from IPython.display import display

import cv2
import numpy as np
from PIL import Image

#Import Image
img = cv2.imread("noisy_image_sample.jpg", cv2.IMREAD_GRAYSCALE)
cv2_imshow(img)
```

```python
#median blur
median = cv2.medianBlur(img, 7)
cv2_imshow(median)
```

```python
#Clahe contasting
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
clahe_img = clahe.apply(median)
cv2_imshow(clahe_img)
```

```python
#OTSU Thresholding
_, thresh_otsu = cv2.threshold(clahe_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
cv2_imshow(thresh_otsu)
```

```python
cv2.imwrite("processed_image.jpg", thresh_otsu)
```
