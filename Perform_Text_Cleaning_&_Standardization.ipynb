```python
import re
import unicodedata
import ftfy
import contractions
import string

with open('text_sample.txt', 'r', encoding='utf-8') as f:
    content = f.read()

fixed_text = ftfy.fix_text(content)
normalized_text = unicodedata.normalize('NFKD', fixed_text).encode('ascii', 'ignore').decode()
expanded_text = contractions.fix(normalized_text)

# standardize before stripping punctuation
common_terms = {
    "USD": "dollars",
    "$": "dollars"
}

for key, value in common_terms.items():
    expanded_text = expanded_text.replace(key, value)

# remove HTML tags
clean_text = re.sub(r'<[^>]+>', ' ', expanded_text)

# remove emails and phone numbers
clean_text = re.sub(r'\S+@\S+', ' ', clean_text)
clean_text = re.sub(r'\(?\d{3}\)?[-\s]?\d{3}[-\s]?\d{4}', ' ', clean_text)

# remove repeated punctuation/symbols
clean_text = re.sub(r'[^\w\s.]', ' ', clean_text)

# remove extra spaces
clean_text = re.sub(r'\s+', ' ', clean_text).strip()

print(clean_text)
```
